```python
import pandas as pd
import pylab as p
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import ExtraTreesClassifier
from sklearn import cross_validation
from sklearn.grid_search import GridSearchCV
import scipy
import random
import time
import requests
from requests.adapters import HTTPAdapter
from multiprocessing import Pool
```

```python
def get_combine_stats(url, columns=None):
    all_df = None
    for year in range(1999, 2014):
        row_check = None
        print "Working on year: {0}".format(year)
        
        session = requests.Session()
        session.mount('http://nflcombineresults.com', HTTPAdapter(max_retries=10))
        r = session.get(url.format(year), timeout=5.0,)
        dfs = pd.read_html(r.content, infer_types=False, header=0, attrs = {'class': 'sortable'})
        if columns is not None:
            dfs[0].columns = columns
        print dfs[0]
        if all_df is None:
            all_df = dfs[0]
        else:
            all_df = all_df.append(dfs[0])
        time.sleep(random.randint(10, 30))
    return all_df
```

```python
url ="http://nflcombineresults.com/nflcombinedata.php?year={0}&pos=&college="
columns = ['Year','Name','College','POS','Height','Weight','Wonderlic','40','BenchPress','VertLeap','BroadJump','Shuttle','3Cone']
combine_df = get_combine_stats(url, columns=columns)
```
