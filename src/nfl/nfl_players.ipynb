```python
def categorical_to_integer(column, df, column_to_integer=None):
    grouped = df.groupby(column)
    if column_to_integer is None:
        column_to_integer = {}
        for i, (name, group) in enumerate(grouped):
            column_to_integer[name] = i
    df['{0}_to_integer'.format(column)] = df[column].apply(lambda x: column_to_integer[x] if str(x) != 'nan' else np.nan)
    return df, column_to_integer
```

```python
# offense_df = offense_df.join(win_orig_df['winner'])
def get_winner(ix, row):
    orig_row = game_df.ix[ix]
#     print row
#     print orig_row
    if orig_row['H'] == row['TEAM'] and orig_row['PTSH'] > orig_row['PTSV']:
        return 1
    if orig_row['H'] == row['TEAM'] and orig_row['PTSH'] < orig_row['PTSV']:
        return -1
    if orig_row['V'] == row['TEAM'] and orig_row['PTSV'] > orig_row['PTSH']:
        return 1
    if orig_row['V'] == row['TEAM'] and orig_row['PTSV'] < orig_row['PTSH']:
        return -1
    else:
        return 0

def get_spread(ix, row):
    orig_row = game_df.ix[ix]
    if orig_row['H'] == row['TEAM']:
        return orig_row['PTSH'] - orig_row['PTSV']
    elif orig_row['V'] == row['TEAM']:
        return orig_row['PTSV'] - orig_row['PTSH']
    else:
        return -1
    
winner_s = []
for ix, row in offense_df.iterrows():
    winner = get_winner(ix, row)
    spread = get_spread(ix, row)
    winner_s.append({"index":row['UID'], "winner":winner, "spread":spread})
join_df = pd.DataFrame(winner_s)
join_df.set_index('index', inplace=True, verify_integrity=False)
offense_df['GID'] = offense_df.index
offense_df.set_index('UID', inplace=True, verify_integrity=False)
offense_df = offense_df.join(join_df)

winner_s = []
for ix, row in defense_df.iterrows():
    winner = get_winner(ix, row)
    spread = get_spread(ix, row)
    winner_s.append({"index":row['UID'], "winner":winner, "spread":spread})
join_df = pd.DataFrame(winner_s)
join_df.set_index('index', inplace=True, verify_integrity=False)
defense_df['GID'] = defense_df.index
defense_df.set_index('UID', inplace=True, verify_integrity=False)
defense_df = defense_df.join(join_df)
#offense_df.set_index("PLAYER", inplace=True, verify_integrity=False)
```

```python
stats_names = ['size', 'min_max', 'mean', 'variance', 'skewness', 'kurtosis']
def push_stats(stats, dicti, formatter):
    for i, stat in enumerate(stats):
        if i != 1:
            dicti[formatter.format(stats_names[i])] = stat
        else:
            dicti[formatter.format('min')] = stat[0]
            dicti[formatter.format('max')] = stat[1]

stats_template = (np.nan, (np.nan, np.nan), np.nan, np.nan, np.nan, np.nan)
            
def get_blocks(player_ids, blocks_df, players_agg_row, all_play_ids):
    block_player = blocks_df[blocks_df['BLK'].isin(player_ids)]
    players_agg_row['num_blocks'] = len(block_player)
    block_player = blocks_df[blocks_df['BRCV'].isin(player_ids)]
    players_agg_row['num_blocks_recovered'] = len(block_player)
    all_play_ids.append(block_player['PID'])

def get_conversions(player_ids, convs_df, players_agg_row, all_play_ids):
    conv_player = convs_df[convs_df['BC'].isin(player_ids)]
    players_agg_row['num_successful_conversion_running'] = len(conv_player[conv_player['CONV'] == 'Y'])
    players_agg_row['num_failed_conversion_running'] = len(conv_player[conv_player['CONV'] == 'N'])
    conv_player = convs_df[convs_df['PSR'].isin(player_ids)]
    players_agg_row['num_successful_conversion_passer'] = len(conv_player[conv_player['CONV'] == 'Y'])
    players_agg_row['num_failed_conversion_passer'] = len(conv_player[conv_player['CONV'] == 'N'])
    conv_player = convs_df[convs_df['TRG'].isin(player_ids)]
    players_agg_row['num_successful_conversion_receiver'] = len(conv_player[conv_player['CONV'] == 'Y'])
    players_agg_row['num_failed_conversion_receiver'] = len(conv_player[conv_player['CONV'] == 'N'])
    all_play_ids.append(conv_player['PID'])

def get_field_goals(player_ids, fgxp_df, players_agg_row, all_play_ids):
    field_goal_player = fgxp_df[fgxp_df['FKICKER'].isin(player_ids)]
    players_agg_row['num_successful_fg'] = len(field_goal_player[(field_goal_player['GOOD'] == 'Y') & (field_goal_player['FGXP'] == 'FG')])
    players_agg_row['num_successful_xp'] = len(field_goal_player[(field_goal_player['GOOD'] == 'Y') & (field_goal_player['FGXP'] == 'XP')])
    players_agg_row['num_failed_fg'] = len(field_goal_player[(field_goal_player['GOOD'] == 'N') & (field_goal_player['FGXP'] == 'FG')])
    players_agg_row['num_failed_xp'] = len(field_goal_player[(field_goal_player['GOOD'] == 'N') & (field_goal_player['FGXP'] == 'XP')])
    if len(field_goal_player['DIST']) > 0:
        stats = scipy.stats.describe(field_goal_player['DIST'])
    else:
        stats = stats_template
    push_stats(stats, players_agg_row, "{0}_field_goal_dist")
    all_play_ids.append(field_goal_player['PID'])

def get_fumbles(player_ids, fumbles_df, players_agg_row, all_play_ids):
    fumbles_player = fumbles_df[fumbles_df['FUM'].isin(player_ids)]
    players_agg_row['num_fumbles'] = len(fumbles_player)
    fumbles_player = fumbles_df[fumbles_df['FRCV'].isin(player_ids)]
    players_agg_row['num_fumbles_recovered'] = len(fumbles_player)
    if len(fumbles_player['FRY']) > 0:
        stats = scipy.stats.describe(fumbles_player['FRY'])
    else: 
        stats = stats_template
    push_stats(stats, players_agg_row, "{0}_fumble_return_ydg")
    fumbles_player = fumbles_df[fumbles_df['FORC'].isin(player_ids)]
    players_agg_row['num_fumbles_forced'] = len(fumbles_player)
    all_play_ids.append(fumbles_player['PID'])

def get_interceptions(player_ids, ints_df, players_agg_row, all_play_ids):
    int_player = ints_df[ints_df['INT'].isin(player_ids)]
    players_agg_row['num_interceptions'] = len(int_player)
    if len(int_player['IRY']) > 0:
        stats = scipy.stats.describe(int_player['IRY'])
    else:
        stats = stats_template
    push_stats(stats, players_agg_row, "{0}_interceptions_ydg")
    players_agg_row['total_interceptions_ydg'] = np.sum(int_player['IRY'])
    all_play_ids.append(int_player['PID'])

def get_kickoffs(player_ids, kickoffs_df, players_agg_row, all_play_ids):
    kickoffs_player = kickoffs_df[kickoffs_df['KICKER'].isin(player_ids)]
    players_agg_row['num_kickoffs'] = len(kickoffs_player)
    if len(kickoffs_player['KGRO']) > 0:
        stats = scipy.stats.describe(kickoffs_player['KGRO'])
    else:
        stats = stats_template
    push_stats(stats, players_agg_row, "{0}_kickoffs_gross")
    if len(kickoffs_player['KNET']) > 0:
        stats = scipy.stats.describe(kickoffs_player['KNET'])
    else:
        stats = stats_template
    push_stats(stats, players_agg_row, "{0}_kickoffs_net")
    players_agg_row['total_kickoffs_net'] = np.sum(kickoffs_player['KNET'])
    players_agg_row['num_kickoffs_touchbacks'] = len(kickoffs_player[kickoffs_player['KTB'] == 'Y'])
    players_agg_row['num_kickoffs_non_touchbacks'] = len(kickoffs_player[kickoffs_player['KTB'] == 'N'])
    all_play_ids.append(kickoffs_player['PID'])
    kickoffs_player = kickoffs_df[kickoffs_df['KR'].isin(player_ids)]
    if len(kickoffs_player['KRY']) > 0:
        stats = scipy.stats.describe(kickoffs_player['KRY'])
    else:
        stats = stats_template
    push_stats(stats, players_agg_row, "{0}_kickoff_kickreturn_ydg")
    players_agg_row['total_kickoff_kickreturn_ydg'] = np.sum(kickoffs_player['KRY'])
    players_agg_row['num_returns'] = len(kickoffs_player)
    all_play_ids.append(kickoffs_player['PID'])

def get_pass(player_ids, pass_df, players_agg_row, all_play_ids):
    pass_player = pass_df[pass_df['PSR'].isin(player_ids)]
    
    players_agg_row['num_passes'] = len(pass_player)
    if len(pass_player['YDS']) > 0:
        stats = scipy.stats.describe(pass_player['YDS'])
    else:
        stats = stats_template
    push_stats(stats, players_agg_row, "{0}_passer_yds")
    players_agg_row['total_passer_yds'] = np.sum(pass_player['YDS'])
    all_play_ids.append(pass_player['PID'])
    
    pass_player = pass_df[pass_df['TRG'].isin(player_ids)]
    players_agg_row['num_receptions'] = len(pass_player)
    if len(pass_player['YDS']) > 0:
        stats = scipy.stats.describe(pass_player['YDS'])
    else:
        stats = stats_template
    push_stats(stats, players_agg_row, "{0}_receiving_yds")
    players_agg_row['total_receiving_yds'] = np.sum(pass_player['YDS'])
    
    location = {"SM":1, "SL":2, "SR":3, "DM":4, "DL":5, "DR":6, "NL":7}
    for key, val in location.iteritems():
        pass_player = pass_df[(pass_df['PSR'].isin(player_ids)) & (pass_df['LOC'] == key)]
        players_agg_row['{0}_num_passes'.format(key)] = len(pass_player)
        if len(pass_player['YDS']) > 0:
            stats = scipy.stats.describe(pass_player['YDS'])
        else:
            stats = stats_template
        push_stats(stats, players_agg_row, "{0}_"+key+"_passer_yds")
        players_agg_row['{0}_total_passer_yds'.format(key)] = np.sum(pass_player['YDS'])
    
    for key, val in location.iteritems():
        pass_player = pass_df[(pass_df['TRG'].isin(player_ids)) & (pass_df['LOC'] == key)]
        players_agg_row['{0}_num_receptions'.format(key)] = len(pass_player)
        if len(pass_player['YDS']) > 0:
            stats = scipy.stats.describe(pass_player['YDS'])
        else:
            stats = stats_template
        push_stats(stats, players_agg_row, "{0}_"+key+"_receiving_yds")
        players_agg_row['{0}_total_receiving_yds'.format(key)] = np.sum(pass_player['YDS'])
    
#    no_huddle_player = pass_player[pass_player['PID'].any(no_huddle_df.index)]
#    players_agg_row['num_no_huddle_passes'] = len(no_huddle_player)
    
    all_play_ids.append(pass_player['PID'])

def get_penalties(player_ids, penalties_df, players_agg_row, all_play_ids):
    penalties_player = penalties_df[penalties_df['PEN'].isin(player_ids)]
    players_agg_row['num_penalties'] = len(penalties_player)
    if len(penalties_player['PEY']) > 0:
        stats = scipy.stats.describe(penalties_player['PEY'])
    else:
        stats = stats_template
    push_stats(stats, players_agg_row, "{0}_penalty_yds")
    players_agg_row['total_penalty_yds'] = np.sum(penalties_player['PEY'])
    all_play_ids.append(penalties_player['PID'])

def get_punts(player_ids, punts_df, players_agg_row, all_play_ids):
    punts_player = punts_df[punts_df['PUNTER'].isin(player_ids)]
    players_agg_row['num_punts'] = len(punts_player)
    if len(punts_player['PGRO']) > 0:
        stats = scipy.stats.describe(punts_player['PGRO'])
    else:
        stats = stats_template
    push_stats(stats, players_agg_row, "{0}_gross_punt_yds")
    players_agg_row['total_gross_punt_yds'] = np.sum(punts_player['PGRO'])
    if len(punts_player['PNET']) > 0:
        stats = scipy.stats.describe(punts_player['PNET'])
    else:
        stats = stats_template
    push_stats(stats, players_agg_row, "{0}_net_punt_yds")
    players_agg_row['total_net_punt_yds'] = np.sum(punts_player['PNET'])
    players_agg_row['num_punts_touchback'] = len(punts_player[punts_player['PTB'] == 'Y'])
    players_agg_row['num_punts_non_touchback'] = len(punts_player[punts_player['PTB'] == 'N'])
    all_play_ids.append(punts_player['PID'])
    punts_player = punts_df[punts_df['PR'].isin(player_ids)]
    players_agg_row['num_punts_recieved'] = len(punts_player)
    if len(punts_player['PRY']) > 0:
        stats = scipy.stats.describe(punts_player['PRY'])
    else:
        stats = stats_template
    push_stats(stats, players_agg_row, "{0}_punt_return_yds")
    players_agg_row['total_punt_return_yds'] = np.sum(punts_player['PRY'])
    players_agg_row['num_punts_fair_caught'] = len(punts_player[punts_player['PFC'] == 'Y'])
    all_play_ids.append(punts_player['PID'])

def get_rush(player_ids, rush_df, players_agg_row, all_play_ids):
    rush_player = rush_df[rush_df['BC'].isin(player_ids)]
    players_agg_row['num_rushes'] = len(rush_player)
    if len(rush_player['YDS']) > 0:
        stats = scipy.stats.describe(rush_player['YDS'])
    else:
        stats = stats_template
    push_stats(stats, players_agg_row, "{0}_rush_yds")
    players_agg_row['total_rush_yds'] = np.sum(rush_player['YDS'])
    all_play_ids.append(rush_player['PID'])

def get_sacks(player_ids, sacks_df, players_agg_row, all_play_ids):
    sacks_player = sacks_df[sacks_df['QB'].isin(player_ids)]
    players_agg_row['num_sacked'] = len(sacks_player)
    if len(sacks_player['YDSL']) > 0:
        stats = scipy.stats.describe(sacks_player['YDSL'])
    else:
        stats = stats_template
    push_stats(stats, players_agg_row, "{0}_sacked_yds")
    players_agg_row['total_sacked_yds'] = np.sum(sacks_player['YDSL'])
    all_play_ids.append(sacks_player['PID'])
    
    sacks_player = sacks_df[sacks_df['SK'].isin(player_ids)]
    players_agg_row['num_sacks'] = np.sum(sacks_player['VALUE'])
    if len(sacks_player['YDSL']) > 0:
        stats = scipy.stats.describe(sacks_player['YDSL'])
    else:
        stats = stats_template
    push_stats(stats, players_agg_row, "{0}_sack_yds")
    players_agg_row['total_sack_yds'] = np.sum(sacks_player['YDSL'])
    all_play_ids.append(sacks_player['PID'])

def get_safeties(player_ids, safeties_df, players_agg_row, all_play_ids):
    safeties_player = safeties_df[safeties_df['SAF'].isin(player_ids)]
    players_agg_row['num_safeties'] = len(safeties_player)
    all_play_ids.append(safeties_player['PID'])

def get_tackles(player_ids, tackles_df, players_agg_row, all_play_ids):
    tackles_player = tackles_df[tackles_df['TCK'].isin(player_ids)]
    players_agg_row['num_tackles'] = np.sum(tackles_player['VALUE'])
    all_play_ids.append(tackles_player['PID'])

def get_offense(player_ids, offense_df, players_agg_row, all_play_ids):
    #offense_df
    offense_player = offense_df[offense_df['PLAYER'].isin(player_ids)]
    # offense_player is all the games that a player played in, grouped by id would be all the games a player has played in to this point
    players_agg_row['number_offensive_games_played_in'] = len(offense_player.GID.unique())
    players_agg_row['number_touchdowns_rushing'] = np.sum(offense_player['TDR'])
    players_agg_row['number_touchdowns_receiving'] = np.sum(offense_player['TDRE'])
    players_agg_row['player_offensive_game_number'] = np.max(offense_player['GAME'])
    players_agg_row['seasons_played_offense'] = np.max(offense_player['SEAS'])
    if len(offense_player['TDR']) > 0:
        stats = scipy.stats.describe(offense_player['TDR'])
    else:
        stats = stats_template
    push_stats(stats, players_agg_row, "{0}_touchdowns_rushing_per_game")
    if len(offense_player['TDRE']) > 0:
        stats = scipy.stats.describe(offense_player['TDRE'])
    else:
        stats = stats_template
    push_stats(stats, players_agg_row, "{0}_touchdowns_receiving_per_game")

def get_dbacks(player_ids, dbacks_df, players_agg_row, all_play_ids):
    dbacks_player = dbacks_df[dbacks_df['DFB'].isin(player_ids)]
    players_agg_row['number_passes_defended'] = len(dbacks_player)

def get_defense(player_ids, defense_df, players_agg_row, all_play_ids):
    defense_player = defense_df[defense_df['PLAYER'].isin(player_ids)]
    # offense_player is all the games that a player played in, grouped by id would be all the games a player has played in to this point
    players_agg_row['number_defensive_games_played_in'] = len(defense_player.GID.unique())
    players_agg_row['number_touchdowns_defense'] = np.sum(defense_player['TDD'])
    players_agg_row['player_defensive_game_number'] = np.max(defense_player['GAME'])
    players_agg_row['seasons_played_defense'] = np.max(defense_player['SEAS'])
    if len(defense_player['TDD']) > 0:
        stats = scipy.stats.describe(defense_player['TDD'])
    else:
        stats = stats_template
    push_stats(stats, players_agg_row, "{0}_defensive_touchdowns_per_game")

def get_kickers(player_ids, kickers_df, players_agg_row, all_play_ids):
    kickers_player = kickers_df[kickers_df['PLAYER'].isin(player_ids)]
    players_agg_row['number_kicker_games_played_in'] = len(kickers_player.GID.unique())
    players_agg_row['number_pat'] = np.sum(kickers_player['PAT'])
    players_agg_row['number_short_field_goals'] = np.sum(kickers_player['FGS'])
    players_agg_row['number_medium_field_goals'] = np.sum(kickers_player['FGM'])
    players_agg_row['number_long_field_goals'] = np.sum(kickers_player['FGL'])
    if len(kickers_player['PAT']) > 0:
        stats = scipy.stats.describe(kickers_player['PAT'])
    else:
        stats = stats_template
    push_stats(stats, players_agg_row, "{0}_pats_per_game")
    if len(kickers_player['FGS']) > 0:
        stats = scipy.stats.describe(kickers_player['FGS'])
    else:
        stats = stats_template
    push_stats(stats, players_agg_row, "{0}_fgs_per_game")
    if len(kickers_player['FGM']) > 0:
        stats = scipy.stats.describe(kickers_player['FGM'])
    else:
        stats = stats_template
    push_stats(stats, players_agg_row, "{0}_fgm_per_game")
    if len(kickers_player['FGL']) > 0:
        stats = scipy.stats.describe(kickers_player['FGL'])
    else:
        stats = stats_template
    push_stats(stats, players_agg_row, "{0}_fgl_per_game")


def aggregate_player(player_id, filtered_play_id_dfs, filtered_game_id_dfs):
    blocks_df, convs_df, fgxp_df, fumbles_df, ints_df, kickoffs_df, pass_df, penalties_df, punts_df, rush_df, sacks_df, safeties_df, tackles_df, dbacks_df = filtered_play_id_dfs
    offense_df, defense_df, core_df, kickers_df = filtered_game_id_dfs
    ix = [player_id]
    row = players_df.ix[ix[0]]
    all_play_ids = pd.Series()
    #print "Got rows"
    players_agg_row = {"player_id":ix, "POS1_to_integer":row['POS1_to_integer'], "HEIGHT":row['HEIGHT'], "WEIGHT":row['WEIGHT'], "YOB":row['YOB'], "FORTY":row['FORTY'], "BENCH":row['BENCH'], "VERTICAL":row['VERTICAL'], "BROAD":row['BROAD'], "SHUTTLE":row['SHUTTLE'], "CONE":row['CONE'], "DPOS":row['DPOS'], "COL_to_integer":row['COL_to_integer'], "DV_to_integer":row["DV_to_integer"], "START":row['START'] }
    #Blocks
    get_blocks(ix, blocks_df, players_agg_row, all_play_ids)
    #print "Done Blocks"
    #Conversions
    get_conversions(ix, convs_df, players_agg_row, all_play_ids)
    #print "Done Conversions"
    #Field Goals
    get_field_goals(ix, fgxp_df, players_agg_row, all_play_ids)
    #print "Done Field Goals"
    #Fumbles
    get_fumbles(ix, fumbles_df, players_agg_row, all_play_ids)
    #print "Done Fumbles"
    #Interceptions
    get_interceptions(ix, ints_df, players_agg_row, all_play_ids)
    #print "Done Intersception"
    #KICKERS
    get_kickers(ix, kickers_df, players_agg_row, all_play_ids)
    #KICKOFFS
    get_kickoffs(ix, kickoffs_df, players_agg_row, all_play_ids)
    #print "Done Kickoffs"
    #PASS
    get_pass(ix, pass_df, players_agg_row, all_play_ids)
    #print "Done pass"
    #Penalties
    get_penalties(ix, penalties_df, players_agg_row, all_play_ids)
    #print "Done penalties"
    #Punts
    get_punts(ix, punts_df, players_agg_row, all_play_ids)
    #print "Done PUnts"
    #RUSHES
    get_rush(ix, rush_df, players_agg_row, all_play_ids)
    #print "Done Rushes"
    #SACKS
    get_sacks(ix, sacks_df, players_agg_row, all_play_ids)
    #print "Done sacks"
    #SAFETIES
    get_safeties(ix, safeties_df, players_agg_row, all_play_ids)
    #print "Done safeties"
    #TACKLES
    get_tackles(ix, tackles_df, players_agg_row, all_play_ids)
    #print "Done Tackles"
    #Offense
    get_offense(ix, offense_df, players_agg_row, all_play_ids)
    #print "Done touchdowns"
    #Dbacks
    get_dbacks(ix, dbacks_df, players_agg_row, all_play_ids)
    # Defense
    get_defense(ix, defense_df, players_agg_row, all_play_ids)
    players_agg_row['total_plays_played_in'] = len(np.unique(all_play_ids.values))
    return players_agg_row     
```

```python
def aggregate_player_proxy(args):
    ix, filtered_play_id_dfs, filtered_game_id_dfs = args
    return aggregate_player(ix, filtered_play_id_dfs, filtered_game_id_dfs)
```

```python
def generate_team_df(df):
    grouped = df.groupby(["GID", "TEAM"])
    ignore_list = ['winner', 'spread', 'GID', 'player_id', 'POS1', 'POS2', 'POS1_to_integer', 'PNAME','FNAME', 'LNAME', 'COL', 'CTEAM', 'DV']#, 'COL_to_integer',, 'FORTY','BENCH','VERTICAL','BROAD','SHUTTLE','CONE', 'DPOS']
    flattened = []
    pool = Pool(processes=8)
    pool_queue = []
    # one group for each game/team combo
    for j, ((game_id, team), group) in enumerate(grouped):
        flattened_dict = {'GID':game_id, 'TEAM':team}
        play_id_dfs = [blocks_df, convs_df, fgxp_df, fumbles_df, ints_df, kickoffs_df, pass_df, penalties_df, punts_df, rush_df, sacks_df, safeties_df, tackles_df, dbacks_df]
        game_id_dfs = [offense_df, defense_df, core_df, kickers_df]
        filtered_play_id_dfs = []
        filtered_game_id_dfs = []
        up_to_play_id = core_df[core_df['GID'] == game_id]['PID'].min()
        for df in play_id_dfs:
            filtered_play_id_dfs.append(df[df['PID'] < up_to_play_id])
        for df in game_id_dfs:
            filtered_game_id_dfs.append(df[df['GID'] < game_id])
        # Loop through each player
    #     print len(group)
        players = []
        pool_queue = []
        for i, (ix, row) in enumerate(group.iterrows()):
            #print row
            player_id = row['PLAYER']
            #flattened_dict['winner'] = row['winner']
            #player_aggregate_dict = aggregate_player(ix, filtered_play_id_dfs, filtered_game_id_dfs)
            pool_queue.append([player_id, filtered_play_id_dfs, filtered_game_id_dfs])
        results = pool.map(aggregate_player_proxy, pool_queue)    
        sub_id_dict = {}
        for player_aggregate_dict in results:
            #flattened.append(player_aggregate_dict)
            # if we have more than one player in each position, offset the position integers
            if player_aggregate_dict['POS1_to_integer'] not in sub_id_dict:
                sub_id_dict[player_aggregate_dict['POS1_to_integer']] = 0
            else:
                sub_id_dict[player_aggregate_dict['POS1_to_integer']] = sub_id_dict[player_aggregate_dict['POS1_to_integer']] + 1
            # the id should be 
            player_aggregate_dict['POS1_to_integer'] = "{0}_{1}".format(player_aggregate_dict['POS1_to_integer'], sub_id_dict[player_aggregate_dict['POS1_to_integer']])
            for key, val in player_aggregate_dict.iteritems():
                if "{0}_{1}".format(player_aggregate_dict['POS1_to_integer'], key) not in flattened_dict and key not in ignore_list:
                    flattened_dict["{0}_{1}".format(player_aggregate_dict['POS1_to_integer'], key)] = val
                # we already have a player at this position, add the two results
                elif "{0}_{1}".format(player_aggregate_dict['POS1_to_integer'], key) in flattened_dict and not np.isnan(val):
                    flattened_dict["{0}_{1}".format(player_aggregate_dict['POS1_to_integer'], key)] = flattened_dict["{0}_{1}".format(player_aggregate_dict['POS1_to_integer'], key)]  + val
            if 'winner' not in flattened_dict:
                flattened_dict['winner'] = row['winner']
            if 'spread' not in flattened_dict:
                flattened_dict['spread'] = row['spread']
        if j%1 == 0:
            print "Finished {0}/{1}".format(j, len(grouped))
    #     print flattened_dict
        flattened.append(flattened_dict)
    game_players_df = pd.DataFrame(flattened)
    #game_players_df.set_index(["GID", "TEAM"], inplace=True)
    return game_players_df
```

```python
def bin_by(x, y, nbins=30):
    """
    Bin x by y.
    Returns the binned "x" values and the left edges of the bins
    """
    bins = np.linspace(y.min(), y.max(), nbins+1)
    # To avoid extra bin for the max value
    bins[-1] += 1 

    indicies = np.digitize(y, bins)

    mean = []
    std = []
    for i in xrange(1, len(bins)):
        mean.append(np.mean(x[indicies==i]))
        std.append(np.std(x[indicies==i]))

    # Just return the left edges of the bins
    bins = bins[:-1]

    return mean, std, bins
```

```python
def plot_feature(feature, xlabel=None, ylabel="Win Probability"):
    probs = cfr.predict_proba(x_test[x_test[feature] > -10000])
    print len(probs),len(x_test[x_test[feature] > -10000])
    mean, std, bins = bin_by(probs[:,1], x_test[x_test[feature] > -10000][feature].values, nbins=10)
    plt.plot(bins, mean, 'k-')
    plt.fill_between(bins, np.array(mean)-std, np.array(mean)+std, facecolor='#eeefff', edgecolor="none")
    if xlabel is not None:
        plt.xlabel(xlabel)
    if ylabel is not None:
        plt.ylabel(ylabel)
    plt.title("{0} vs {1}".format(xlabel, ylabel))
    plt.show()    
```
